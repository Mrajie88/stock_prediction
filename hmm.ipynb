```python
import numpy as np
import pandas as pd
from hmmlearn import hmm
```

```python
# 读取数据
filepath = "raw_price_train/1_r_price_train.csv"
data = pd.read_csv(filepath)
data.head()
```

```python
close = data["Adj Close"].values
```

```python
close = close.reshape(-1,1)
```

```python
close
```

```python
n = 5
model = hmm.GaussianHMM(n_components=n,n_iter=2000).fit(close)
```

```python
hidden_states = model.predict(close[-5:])
```
